# Make lineplot pan and core genes 

```python
#load packages
import pandas as pd
import os
os.getcwd()
```

```python
#load data
pan = pd.read_csv("new_pan_genes.csv")
core = pd.read_csv("new_core_genes.csv")
display(pan)
display(core)
```

```python
#melt columns
#https://towardsdatascience.com/wide-to-long-data-how-and-when-to-use-pandas-melt-stack-and-wide-to-long-7c1e0f462a98
pan1 = pan.melt(var_name = "number of genomes", value_name = "number of pan-genes")
core1 = core.melt(var_name = "number of genomes", value_name = "number of core-genes")
display(pan1)
display(core1)
```

```python
#merge columns
#https://stackoverflow.com/questions/19125091/pandas-merge-how-to-avoid-duplicating-columns
cols_to_use = core1.columns.difference(pan1.columns) #have to use this first before pd.merge to avoid duplicated values
df = pd.merge(pan1, core1[cols_to_use], left_index=True, right_index=True, how='outer')
df.head(11)
```

```python
#write to csv file
df.to_csv('new_pan-core-genes.csv', index=False)
```

```python
#import pan-core-genes data
import seaborn as sns
```

```python
#making plot
#https://seaborn.pydata.org/generated/seaborn.lineplot.html
pan_core_long = df.melt('number of genomes', var_name = "genomes", value_name = "number of genes")
#g = sns.catplot(x="number of genomes", y="number of genes", hue='genes cluster', data=pan_core_long, kind='point')
sns.set(rc={"figure.figsize":(13,8)})
sns.lineplot(
    data=pan_core_long, x="number of genomes", y="number of genes", hue="genomes", err_style="bars", ci=80
)
```
